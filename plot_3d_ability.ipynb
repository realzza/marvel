```python
import plotly.io as pio
pio.renderers.default = 'colab' # or 'notebook' or 'colab' or 'jupyterlab'
import pandas as pd
```

```python
df_marvel = pd.read_csv('superheroes/charcters_stats.csv')
```

```python
df_marvel.head()
```

```python
df_marvel = df_marvel.dropna()
```

```python
import plotly.express as px
# df = px.data.iris()
fig = px.scatter_3d(df_marvel, x='Intelligence', y='Durability', z='Combat', title="Marvel Characters and Corresponding Power",
              color='Total',symbol="Alignment",hover_name="Name",opacity=0.7,color_continuous_scale="Picnic")
fig.show()
```

```python
fig.update_layout(coloraxis_colorbar=dict(yanchor="top", y=1, x=0))
```

```python
fig.update_layout(
    scene = dict(
        xaxis = dict(
            gridcolor="rgb(50,50,50)",
            backgroundcolor="rgb(0,0,0)",
            showbackground=True,
        ),
        yaxis = dict(
            gridcolor="rgb(50,50,50)",
            backgroundcolor="rgb(0,0,0)",
            showbackground=True,
        ),
        zaxis = dict(
            gridcolor="rgb(50,50,50)",
            backgroundcolor="rgb(0,0,0)",
            showbackground=True,
        ),
    ),
    font_family = "Avenir",
    font_color="white",
    paper_bgcolor='rgba(0,0,0)',
)
```

```python
fig.update_traces(marker=dict(size=8,
                              line=dict(width=1,
                                        color='White')),
                  selector=dict(mode='markers'))
```
